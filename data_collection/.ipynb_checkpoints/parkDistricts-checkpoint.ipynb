## SCORE TIMESERIES

## Read Calls

```python
calls = pd.read_csv(PARQA + '/data/311/calls_matched_districts.csv',index_col=0)
```

```python
calls.head()
```

## Aggregating Funct Area per district

```python
calls['Created Date'] = pd.to_datetime(calls['Created Date']) # 
```

```python
calls['year'] = calls['Created Date'].dt.year
```

```python
calls.year.value_counts()
```

```python
calls.columns
```

```python
calls1 = calls.groupby('parkDistrict').size().reset_index()
calls1.head()
```

```python
CallsYear = calls.groupby(['parkDistrict','year']).size().unstack().reset_index()
finalCalls = calls1.merge(CallsYear, on='parkDistrict', how='left').rename(columns={0:'total','parkDistrict':'SYSTEM'})
```

```python
finalCalls.head()
```

```python
distr2 = distr.merge(finalCalls, how='left', on='SYSTEM')
```

```python
distr2.head(2)
```

## Read Scores

```python
scores = pd.read_csv(PARQA + 'parqa/311/TIMESERIES/PIP_score_timeseries.csv',index_col=0)
scores = scores[scores.score_type == 'Amenities & Area Normalized Score'].drop('score_type',1)
scores.columns = ['SYSTEM'] + ['score%d'% i for i in xrange(2004,2016)]
```

```python
scores.head()
```

```python
distr3 = distr2.merge(scores, how='left', on='SYSTEM')
```

## SAVE to geojson

```python
path = PARQA + '/Parqa_tool/data/districts.json'

with open(path,'w') as f:
    f.write(distr3.to_json())
```

## Save to topojson

```python
from topojson import topojson

#give it a path in and out
topojson(path,path.replace('.json','_t.json'), quantization=1e6, simplify=0.0001)
```
